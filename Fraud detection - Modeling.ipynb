```
#Importamos librerias
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
```

```
#Reading fraud_df
fraud_dfs = pd.read_csv('fraud_sample.csv')
```

```
#Dividiendo la data en X y Y
X = fraud_dfs.drop('Class', axis=1)
y = fraud_dfs['Class']
```

```
#Dividimos la data en en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```
#Probamos Logistic Regression. Me refiero al training score como accuracy
LogisticRegression().fit(X_train, y_train)
score = cross_val_score(LogisticRegression(), X_train, y_train, cv=5)
print('Logistic Regression tiene un accuracy de: ', round(score.mean(), 2)*100, '%')
```

```
#Probamos Support Vector Classifier. Me refiero al trainig score como acccuracy
SVC().fit(X_train, y_train)
score = cross_val_score(SVC(), X_train, y_train, cv=5)
print('Suppert Vector Classifier tiene un accuracy de: ', round(score.mean(), 2)*100, '%')
```

Logistic reggression y SVC tuvieron el mismo training score, sin embargo logistic regression fue mucho más eficiente porque tardó significativamente menos tiempo. Por eso haremos el test con el modelo resultante de la Logistic Regression

```
#Entrenamos el modelo lr_model
lr_model = LogisticRegression().fit(X_train, y_train)
```

```
#Testing lr_model
y_predict = lr_model.predict(X_test)
```

```
#Confusssion matrix
confusion_matrix(y_test, y_predict)
```
